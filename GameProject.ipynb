```python
import random

# Create players with a name and a point value
class player:
  def __init__(self, name):
    self.name = name
    self.points = 0

  def add_point(self):
    self.points += 1

# Print rules if requested
def print_rules():
  print(" ")
  print("Okay, here are the rules!")
  print("Player 1 selects 2 cards. ")
  print("If the cards match, the player keeps the cards and tries again.")
  print("If the cards do not match, the selected cards will be flipped back over and the next player goes.")
  print("The player with the most matching cards wins the game.")
  print(" ")

# Returns a list of cards based on how many cards the players want and which side of the card it is (front or back)
def create_deck(num, side):
  rtn_list = []
  front_deck = ["AS","2S", "3S", "4S", "5S","6S", "7S", "8S", "9S", "10S","JS", "QS", "KS",
               "AH", "2H", "3H", "4H", "5H","6H", "7H", "8H", "9H", "10H","JH", "QH", "KH",
               "AD", "2D", "3D", "4D", "5D","6D", "7D", "8D", "9D", "10D","JD", "QD", "KD",
               "AC", "2C", "3C", "4C", "5C","6C", "7C", "8C", "9C", "10C","JC", "QC", "KC"]

  if side == "front":                   # Create a list of card values using the set values of full_deck and the number of cards the player wants
    for i in range(0,int(num)):
      for x in range(0,2):              # Nested for loop to duplicate each card so its a list of pairs
        rtn_list.append(front_deck[i])
    random.shuffle(rtn_list)            # Shuffle it so all the values are in a random spot

  else:                                 # Create a list of the backs of the cards each given a number
    for i in range(0, (int(num) * 2)):       # Give each turned over card a unique number so the player can choose one  by its number
      rtn_list.append(i)

  return rtn_list

# Prints the scoreboard nicely
def print_points_nice(p1, p2):
  print(p1.name + ": ", str(p1.points) + " pts   |   " + p2.name + ": ", str(p2.points) + " pts ")

# Prints the cards nicely in rows of 16
def print_cards_nice(cards):
  print("--------------------------------------------")
  if len(cards) <= 16:
    for i in range (len(cards)):
      print(cards[i], end = "  ")

  else:
    for i in range (len(cards)):
      if (i % 16 == 0):
        print("")
      print(cards[i], end = "  ")
  print("")
  print("--------------------------------------------")
  print("")

# Count the player's points and return the winner result
def who_won(p1, p2):
  print("")
  if p1.points == p2.points:
    print("Good Game!! It was a tie!")
  elif p1.points > p2.points:
    print("Good Game! ", p1.name + " won!")
  elif p2.points > p1.points:
    print("Good Game! ", p2.name + " won!")
  else:
    print("inconclusive")

# Returns True if game is over and false if it is not
def is_game_over(deck):
  for i in range(0, len(deck)):
    # If there are still unturned cards, the game continues
    if deck[i] != " ":
      return False
    # If all the cards have been turned over, they should all be marked as " " and the game ends
  return True

# Check if the chosen card has been chosen already and return true or false
def check_empty(card, back):
  if back[card] == " ":
    return True
  else:
    return False

# Check if the cards are a match and turn true or false
def corr_match(c1, c2, front, back):
  if front[c1] == front[c2]:
    return True
  else:
    return False

# Update the deck by replacing the chosen cards with " "
def update_deck(c1, c2, front, back):
  back[c1] = " "
  back[c2] = " "

# Return's who's turn it is based on who just went
def change_players(p1, p2, player):
  if player == p1:
    return p2
  else:
    return p1


# Returns False if player can go again, True if the turn is over
def player_takes_turn(player, front, back, p1, p2, cards):
  valid_c1 = False
  valid_c2 = False

  while valid_c1 == False:  # Make sure input is valid
    c1 = input(player.name + ", Choose 1st card: ")
    if c1.isdigit() == False:
      print("Not valid, choose a valid number from the list above.")
    elif int(c1) >= (cards * 2):
      print("Not valid, choose a valid number from the list above.")
    elif back[int(c1)] == " ":
      print("Not valid, choose a valid number from the list above.")
    else:
      c1 = int(c1)
      valid_c1 = True

  while valid_c2 == False:  # Make sure input is valid
    c2 = input(player.name + ", Choose 2nd card: ")
    if c2.isdigit() == False:
      print("Not valid, choose a valid number from the list above.")
    elif int(c2) >= (cards * 2) or int(c1) == int(c2):
      print("Not valid, choose a valid number from the list above.")
    elif back[int(c2)] == " ":
      print("Not valid, choose a valid number from the list above.")
    else:
      valid_c2 = True
      c2 = int(c2)

  print("You chose: " + str(front[c1]) + " and " + str(front[c2]))

  if corr_match(c1, c2, front, back) == True:
    update_deck(c1, c2, front, back)
    player.add_point()
    print_points_nice(p1,p2)
    print("Thats a match!")
    return False
  else:
    print("Not a match sorry!")
    return True

  print("")

def run_game():
  # Check if players needs the rules
  rules = input("Welcome to Memory Match! If you would like the rules, type 'rules', otherwise hit enter! ")
  if rules == "rules":
    print_rules()
  print("Okay great! Lets begin!")

  # Player Chooses the size of the game
  total_num_cards = input("Choose number of pairs (between 4 and 52): ")

  # Make sure the player entered a valid number of cards
  valid_num_cards = False
  while valid_num_cards == False:
    if total_num_cards.isdigit() == False:  # Must be an integer
      total_num_cards = input("Choose an integer between 4 and 52! ")
    elif int(total_num_cards) > 52 or int(total_num_cards) < 4:  # Must be within the given range
      total_num_cards = input("Choose an integer between 4 and 52! ")
    else:
      valid_num_cards = True

  # Initiate Players
  p1 = player(input("Enter player 1 name: "))
  p2 = player(input("Enter player 2 name: "))

  # Initiate the Decks
  front = create_deck(total_num_cards, "front")
  back = create_deck(total_num_cards, "back")

  # Set game conditions
  game_over = False
  turn_over = False
  round_num = 1
  turn = p1

  # Loop for entire game
  while game_over == False:
    for x in range(0,2):
      print("")
      print("Round #" + str(round_num) + ": " + turn.name + "'s turn")
      print_points_nice(p1,p2)

      # Loop for individual turns
      while turn_over == False:
        print_cards_nice(back)
        turn_over = player_takes_turn(turn, front, back, p1,p2, int(total_num_cards))

        if is_game_over(back) == True:  # If all the cards are 'empty' the game ends
          game_over = True
          turn_over = True
          break

        if turn_over == True:   # Otherwise, the turn ends and the players sqitch
          turn_over = False
          round_num += 1
          turn = change_players(p1,p2,turn)

  who_won(p1, p2)

# Loop to play the game until the player quits
new_game_start = False
while new_game_start == False:
  run_game()

  print("")
  new_game = input("Do you want to play again or quit? Enter y/n/q/quit")
  if new_game == "n" or new_game == "q" or new_game == "quit":
    new_game_start = True
    break
  else:
    print("Great! Starting a new game........")
    print("")
```
